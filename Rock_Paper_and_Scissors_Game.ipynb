```python
import random

print("WELCOME\nTo the ROCK PAPER AND SCISSORS GAME\n\nFollow the instructions to play the game effectively:")
print("You will be given with 'three' chances.\nYou should win atleast 'two' of them to win the game.")
print("Enter your choices in 'lowercase alphabets' only as follows:\nrock\npaper\nscissors\n")


def Compare(user , computer):
    res = 0
    if(user == computer):
        res = 0                 #computer wins then res equals to 1
    elif user == "rock":        #user wins then res equals to 2
        if computer == "paper":
            res = 1
        elif computer == "scissors":
            res = 2
    elif user == "paper":
        if computer == "rock":
            res = 2
        elif computer == "scissors":
            res = 1
    elif user == "scissors" :
        if computer == "rock":
            res = 1
        elif computer == "paper":
            res = 2
    
    return res

count = 0
count1 = 0
j=0
i=1
while i<4:
    print("Chance " + str(i) + ":")
    result = -1;
    user_choice = input("Enter your choice:")
    if(user_choice == "rock" or user_choice == "paper" or user_choice == "scissors"):
        options = ["rock" , "paper" , "scissors"]
        computer_choice = random.choice(options)
        print("Computer Choice:" + computer_choice)
        result = Compare(user_choice , computer_choice)
        i += 1
        if(result == 0):
                print("Its a tie")
        elif(result == 1):
                count += 1
                print("Computer won this chance")
        elif(result == 2):
                count1 += 1
                print("You won this chance")
    else:
        if j == 0:
            print("You have entered a incorrect choice\nSo your are given an extra chance\nBut if it repeats, next chance it will given that computer won the chance")
            j += 1
        elif j != 0:
            print("Computer won the game")
            count += 1
            i += 1
    
if(count == count1):
    print("\nIts a tie play another time")
elif(count > count1):
    print("\nComputer won the game\nBetter luck next time\n:)")
elif(count < count1):
    print("\nCongratulations\nYou won the game")
```
